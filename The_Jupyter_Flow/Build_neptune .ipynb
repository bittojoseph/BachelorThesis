# Build Submarine Neptune

## 1.Charging lipo Batteries

### With our current hardware setup you need 3 fully charged batteries to have a smooth drive with Submarine. It can be reduced to 2 if space needs to be saved.

### Steps for charging the LiPo batteries

1. Check how many cells are there in the battery and connect it to the Charger as shown in the below picture
2. Select the battery type by click on the switch "Batt. Type". Here we have "LiPo BATT"
3. Click Start button
4. Click inc button and select "Lipo BALANCE" as shown in picture below. Make sure it's 1.6A 11.1V(3S) for the batteries we are using
5. Click and hold Start button until you see "Confirm"
6. Click Start button again to start charging.
